```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('RL-literature production.txt', 'r') as ff:
    df = pd.read_csv(ff, delimiter='\t', na_filter=True)
```

```python
fig, ax = plt.subplots(2,1, figsize=(8,7))
ax[0].set_title("Annual Scientific Production")
sns.barplot(df.sort_index(ascending=False), ax=ax[0], x="Publication Years", y = "Record Count")
ax[0].tick_params(axis='x', rotation=90)
sns.barplot(df.iloc[25:].sort_index(ascending=False), ax=ax[1], x="Publication Years", y = "Record Count")
ax[1].tick_params(axis='x', rotation=90)

```
